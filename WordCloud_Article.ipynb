```python
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
from newspaper import Article
from wordcloud import STOPWORDS
```

```python
# Insert any article URL to scrape words

article = Article('https://www.eater.com/21518621/indoor-dining-restaurants-safety-risks-covid-19-chefs-waiters-servers')
article.download()
article.parse()
article.text

## Find an image with a white background and well defined edges to use as a mask

mask = np.array(Image.open('C:\\Users\\Main PC\\Desktop\\MultiColored-United-States-Map.jpg'))
mask_colors = ImageColorGenerator(mask)
```

```python
# Define parameters and render

wc = WordCloud(stopwords=STOPWORDS, mask=mask, background_color="white",
               max_words=3000, max_font_size=300, random_state=42,
              ).generate(article.text)

image_colors = ImageColorGenerator(mask)

plt.figure(figsize=[7,7])
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
# Save and name

wc.to_file("cloud.png")
```
