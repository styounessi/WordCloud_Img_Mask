```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from newspaper import Article
from wordcloud import STOPWORDS
from wordcloud import WordCloud, ImageColorGenerator
```

```python
# Insert any article URL

article = Article('https://www.brainpickings.org/2015/10/07/esther-sternberg-stress-relationships/')
article.download()
article.parse()
article.text
```

```python
# Find an image with a white background and well defined edges to use as a mask

mask = np.array(Image.open('C:\\Users\\Main PC\\Desktop\\image.jpg'))
mask_colors = ImageColorGenerator(mask)
```

```python
# Define parameters using WordCloud

wc = WordCloud(stopwords=STOPWORDS, mask=mask, background_color='white',
               max_words=2000, max_font_size=None, random_state=42, min_word_length=2
               ).generate(article.text)

image_colors = ImageColorGenerator(mask)
```

```python
# Render final output

plt.figure(figsize=[15,15])
plt.imshow(wc.recolor(color_func=image_colors), interpolation='bilinear')
plt.axis('off')
plt.show()
```
